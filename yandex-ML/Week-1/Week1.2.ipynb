```python
new_data = pd.read_csv("titanic.csv", usecols = ['Pclass','Fare', 'Sex', 'Age', 'Survived'])
```

```python
new_data
```

```python
new_data.dropna()

new_data.insert(0, 'Gender', new_data['Sex'].map({'female': 0, 'male': 1}).astype(int))
```

```python
new_data = new_data.drop('Sex', axis=1)
new_data = new_data.dropna()
```

```python
new_data = new_data.reset_index(drop=True)
new_data
```

```python
features = new_data[ ['Pclass', 'Fare', 'Age', 'Gender'] ]
testX = new_data[ ['Survived'] ]

clf = DecisionTreeClassifier(random_state=241)
clf.fit(features, testX)
```

```python
importances = clf.feature_importances_
print(importances)
```
